#  Выявление закономерностей, определяющих успешность компьютерных игр для интернет-магазина.

Интернет-магазин «Стримчик» продаёт по всему миру компьютерные игры. Из открытых источников доступны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation). Нужно выявить определяющие успешность игры закономерности. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании.
В датасете данные до 2016 года. Сейчас декабрь 2016 г., нужно спланировать кампанию на 2017-й. Нужно отработать принцип работы с данными. В наборе данных попадается аббревиатура ESRB (Entertainment Software Rating Board) — это ассоциация, определяющая возрастной рейтинг компьютерных игр. ESRB оценивает игровой контент и присваивает ему подходящую возрастную категорию, например, «Для взрослых», «Для детей младшего возраста» или «Для подростков».

* User_Score — оценка пользователей (максимум 10)
* Rating — рейтинг от организации ESRB (англ. Entertainment Software Rating Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию.

```python
# Гистограмма
ax = df_XOne.plot(kind='hist', y='user_score', histtype='step', bins=10, linewidth=5, alpha=0.7, label='XOne')
df_PC.plot(kind='hist', y='user_score', histtype='step', bins=10, linewidth=5, alpha=0.7, label='PC', ax=ax, grid=True, legend=True, figsize=(12,6), title='Частота оценок пользователей тарифов PC и XOne');
```

Нулевая гипотеза: Средние пользовательские рейтинги платформ Xbox One и PC одинаковы.

Альтернативная гипотеза: Средние пользовательские рейтинги платформ Xbox One и PC различны (больше или меньше).

Зададим критический уровень статистической значимости 5%, если p-value окажется меньше него - отвергнем гипотезу. Пользователи исследуемых платформ явно из разных сегментов, показатель std  выборок отличается, предыдущие исследования показали что платформы имеют разный рейтинг в разных регионах, поэтому параметр equal_var укажем False.

```python
alpha = .05
```

```python
results = st.ttest_ind(df_XOne, df_PC, equal_var = False)
print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

### Вывод
Значение p-value показывает: вероятность случайно получить такое или большее различие равно почти 55%, средние пользовательские рейтинги платформ Xbox One и PC очень близки.
```python
# Гистограмма
ax = df_Sports.plot(kind='hist', y='user_score', histtype='step', \
                    bins=10, linewidth=5, alpha=0.7, label='Sports')
df_Action.plot(kind='hist', y='user_score', histtype='step', bins=10, linewidth=5, alpha=0.7, \
               label='Action', ax=ax, grid=True, legend=True, figsize=(12,6), \
               title='Частота оценок пользователей жанров Action и Sports');
```

Нулевая гипотеза: Средние пользовательские рейтинги жанров Action и Sports одинаковы.

Альтернативная гипотеза: Средние пользовательские рейтинги жанров Action и Sports различны (больше или меньше).

Зададим критический уровень статистической значимости 5%, если p-value окажется меньше него - отвергнем гипотезу. Пользователи исследуемых жанров явно из разных сегментов, показатель std  выборок отличается, предыдущие исследования показали что жанры имеют разный рейтинг в разных регионах, поэтому параметр equal_var укажем False.

```python
alpha = .05
```

```python
results = st.ttest_ind(df_Action, df_Sports, equal_var = False)
print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

### Вывод
Значение p-value показывает: вероятность случайно получить такое или большее различие очень мала, на это же указывает и большая разница в средних рейтингах пользователей по разным жанрам.  Это явно слишком маленькая вероятность, чтобы делать вывод об одинаковых средних рейтингах. Наша гипотеза о том что средние пользовательские рейтинги жанров Action и Sports  разные не отвергается.
